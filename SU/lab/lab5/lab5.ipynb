```python
from sklearn import linear_model
from sklearn import neighbors  # KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
def get_metrics(regresor, X, y):
    return mean_absolute_error(y, regresor.predict(X)), mean_squared_error(y, regresor.predict(X))
```

```python
print(dane.shape)
Xregr=dane[:,0:-1]
yregr=dane[:,-1]

metrics = {

}

regr = linear_model.LinearRegression()
regr.fit(Xregr, yregr)
metrics['LinearRegression'] = get_metrics(regr, Xregr, yregr)

regr = neighbors.KNeighborsRegressor()
regr.fit(Xregr, yregr)
metrics['KNeighborsRegressor'] = get_metrics(regr, Xregr, yregr)

regr = DecisionTreeRegressor(max_depth=2)
regr.fit(Xregr, yregr)
metrics['DecisionTreeRegressor'] = get_metrics(regr, Xregr, yregr)

regr = MLPRegressor()
regr.fit(Xregr, yregr)
metrics['MLPRegressor'] = get_metrics(regr, Xregr, yregr)

regr = SVR(kernel='linear')
regr.fit(Xregr, yregr)
metrics['SVR-linear'] = get_metrics(regr, Xregr, yregr)

regr = SVR(kernel='rbf')
regr.fit(Xregr, yregr)
metrics['SVR-rbf'] = get_metrics(regr, Xregr, yregr)

print(metrics)
```

```python
models = list(metrics.keys())
mse_values = [val[0] for val in metrics.values()]
rmse_values = [val[1] for val in metrics.values()]

fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = range(len(models))

bar1 = ax.bar(index, mse_values, bar_width, label='Mean absolute Error', color='salmon')
# bar2 = ax.bar([i + bar_width for i in index], rmse_values, bar_width, label='Mean Squared Error')

ax.set_xlabel('Modele')
ax.set_ylabel('WartoÅ›ci metryki')
ax.set_title('PorÃ³wnanie modeli metrykÄ… M1')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(models, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
print(dane.shape)
Xregr=dane[:,0:-1]
yregr=dane[:,-1]

metrics = {

}

regr = make_pipeline(StandardScaler(), linear_model.LinearRegression())
regr.fit(Xregr, yregr)
metrics['LinearRegression'] = get_metrics(regr, Xregr, yregr)

regr = make_pipeline(StandardScaler(), neighbors.KNeighborsRegressor())
regr.fit(Xregr, yregr)
metrics['KNeighborsRegressor'] = get_metrics(regr, Xregr, yregr)

regr = make_pipeline(StandardScaler(), DecisionTreeRegressor(max_depth=2))
regr.fit(Xregr, yregr)
metrics['DecisionTreeRegressor'] = get_metrics(regr, Xregr, yregr)

regr = make_pipeline(StandardScaler(), MLPRegressor())
regr.fit(Xregr, yregr)
metrics['MLPRegressor'] = get_metrics(regr, Xregr, yregr)

regr = make_pipeline(StandardScaler(), SVR(kernel='linear'))
regr.fit(Xregr, yregr)
metrics['SVR-linear'] = get_metrics(regr, Xregr, yregr)

regr = make_pipeline(StandardScaler(), SVR(kernel='rbf'))
regr.fit(Xregr, yregr)
metrics['SVR-rbf'] = get_metrics(regr, Xregr, yregr)

print(metrics)
```

```python
models = list(metrics.keys())
mse_values = [val[0] for val in metrics.values()]
rmse_values = [val[1] for val in metrics.values()]

fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = range(len(models))

bar1 = ax.bar(index, mse_values, bar_width, label='Mean absolute Error', color='salmon')
# bar2 = ax.bar([i + bar_width for i in index], rmse_values, bar_width, label='Mean Squared Error')

ax.set_xlabel('Modele')
ax.set_ylabel('WartoÅ›ci metryki')
ax.set_title('PorÃ³wnanie modeli z normalizacjÄ… danych metrykÄ… M1')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(models, rotation=45, ha='right')
ax.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
